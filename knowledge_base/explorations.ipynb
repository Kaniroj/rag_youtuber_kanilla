```python
from pathlib import Path
from pypdf import PdfReader

pdf_path = Path("data/Netherland_Dwarf_rabbit.pdf")
reader = PdfReader(pdf_path)

all_text = ""
for page in reader.pages:
    all_text += (page.extract_text() or "") + "\n"

print(all_text[:1000])
len(all_text)
```
